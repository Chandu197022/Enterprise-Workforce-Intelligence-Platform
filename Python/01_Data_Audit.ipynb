```python
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[18,25,35,45,55,65],
    labels=[
        '18-25',
        '26-35',
        '36-45',
        '46-55',
        '55+'
    ]
)

df[['Age','AgeGroup']].head()
```

```python
df['EmployeeExperienceIndex'] = (
    df['JobSatisfaction']
    + df['EnvironmentSatisfaction']
    + df['WorkLifeBalance']
    + df['RelationshipSatisfaction']
)/4

df['EmployeeExperienceIndex'].head()
```

```python
df['RiskScore'] = 0
```

```python
df.loc[
    df['OverTime']=='Yes',
    'RiskScore'
] += 2
```

```python
salary_threshold = df['MonthlyIncome'].median()

df.loc[
    df['MonthlyIncome'] < salary_threshold,
    'RiskScore'
] += 2
```

```python
df.loc[
    df['JobSatisfaction'] <= 2,
    'RiskScore'
] += 2
```

```python
df.loc[
    df['YearsAtCompany'] <= 2,
    'RiskScore'
] += 1
```

```python
df.loc[
    df['BusinessTravel']
    == 'Travel_Frequently',
    'RiskScore'
] += 1
```

```python
def classify_risk(score):

    if score <= 2:
        return 'Low Risk'

    elif score <= 4:
        return 'Medium Risk'

    elif score <= 6:
        return 'High Risk'

    else:
        return 'Critical Risk'
```

```python
def classify_risk(score):

    if score <= 2:
        return 'Low Risk'

    elif score <= 4:
        return 'Medium Risk'

    elif score <= 6:
        return 'High Risk'

    else:
        return 'Critical Risk'
```

```python
df['RiskCategory'] = df['RiskScore'].apply(classify_risk)
```

```python
df[['RiskScore','RiskCategory']].head()
df['RiskCategory'].value_counts()
```

```python
df['RiskCategory'] = (
    df['RiskScore']
    .apply(classify_risk)
)

df[['RiskScore','RiskCategory']].head()
```

```python
df['RiskCategory'].value_counts()
```

```python
df.to_csv(
    "../Dataset/HR_Enhanced.csv",
    index=False
)

print("Enhanced Dataset Saved Successfully")
```
